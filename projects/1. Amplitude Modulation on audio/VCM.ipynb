# <span style="color:rgb(213,80,0)">Amplitude Modulation</span>

```matlab
% Bandwidth of msg_signal (message signal)
%BW = 4khz;
fc = 10000 % to prevent aliasing bez Fs = 48000 so fm <= Fs/2 = 24000hz,
```

```matlab
           % so FC+FM = 14000=AM  so for output without aliasing
           % 2(AM) = 2(14000)= 48000 = Fs
c = cos(2*pi*fc*t)
```

```matlab
plot(t,c)
```

```matlab
size(msg_sig)
```

```matlab
size(c)
```

```matlab
c = c(:)
```

```matlab
size(msg_sig)
```

```matlab
size(c)
```

```matlab
ka = 0.6
```

```matlab

AM_signal = (1 + ka*msg_sig).*c
```

```matlab
figure;
subplot(2, 1, 1)
plot(t, msg_sig);
title("message signal");

subplot(2, 1, 2)
plot(t, AM_signal)
xlim([1, 1.015])
title("AM modulated signal")
```

# <span style="color:rgb(213,80,0)">Am Under noise</span>

```matlab
%% Effect of Different SNR Values

snr_values = [0 5 10 20];

figure('Name','SNR Comparison')

for i = 1:length(snr_values)

    % Additive white gaussion noise
    nAm = awgn(AM_signal,snr_values(i),'measured');

    % envlope detector
    env = abs(hilbert(nAm));

    % low pass filter
    [b,a] = butter(4, 4000/(fs/2));
    n_recovered = filtfilt(b, a, env);

    % removal of dc
    noisy_recovered = n_recovered - mean(n_recovered);

    % Normalize audio
    recovered = noisy_recovered/max(abs(noisy_recovered));

    subplot(2,2,i)

    plot(t,recovered)

    title(['Recovered Signal (SNR = ',num2str(snr_values(i)),' dB)'])
    grid on

    % Play recovered audio
    sound(recovered,fs)

    pause(length(recovered)/fs + 1)

    grid on

end
```
