
# <span style="color:rgb(213,80,0)">Phase Modulation</span>

```matlab
clc;clear;

fs = 10000;
t = .0: 1/fs:0.2;

Am = 1;
fm = 100;

fc = 700;
Ac = 1;

m = Am*cos(2*pi*fm*t);

kp = 3;
spm = Ac*cos(2*pi*fc*t + kp*m);

subplot(3,1,1);
plot(t,m);
title('Message Signal');
grid on;

subplot(3,1,2);
plot(t,spm);
title('PM Signal');
grid on;

subplot(3,1,3);
plot(t(1:1000),spm(1:1000));
title('Zoomed PM Signal');
grid on;
```

```matlab
N = length(t);

f = (-N/2: N/2-1)*(fs/N);

s = fftshift(abs(fft(spm)));
```

```matlab
figure;
plot(f,s);
xlim([fc-1000 fc+1000]);

title('PM Spectrum');
xlabel('Frequency (Hz)');
ylabel('Magnitude');
grid on;
```
