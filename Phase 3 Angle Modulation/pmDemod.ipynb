
# <span style="color:rgb(213,80,0)">Demodulating PM signal</span>

```matlab
clc;clear;

fs = 10000;
t = .0: 1/fs:0.2;

Am = 1;
fm = 100;

fc = 700;
Ac = 1;

m = Am*cos(2*pi*fm*t);

kp = 3;
spm = Ac*cos(2*pi*fc*t + kp*m);

subplot(3,1,1);
plot(t,m);
title('Message Signal');
grid on;

subplot(3,1,2);
plot(t,spm);
title('PM Signal');
grid on;

subplot(3,1,3);
plot(t(1:500),spm(1:500));
title('Zoomed PM Signal');
grid on;
```

# <span style="color:rgb(213,80,0)">Demodulation</span>

```matlab
pm_demod = pmdemod(spm, fc, fs, kp);
figure;
subplot(3,1,1);
plot(t,m);
title('Original Message');
grid on;

subplot(3,1,2);
plot(t,spm);
title('PM Signal');
grid on;

subplot(3,1,3);
plot(t,pm_demod);
title('Recovered Message');
grid on;
```

# <span style="color:rgb(213,80,0)">FFt an̥alysis</span>

```matlab
N = length(t);
f = (-N/2: N/2-1)*(fs/N);

fftsignal = fftshift(abs(fft(pm_demod)));
figure;
plot(f, fftsignal)
xlim([-400, 400])
```
